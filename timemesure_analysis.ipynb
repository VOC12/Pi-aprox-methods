### Comparing the three methods with N=1000

```python
import matplotlib.pyplot as plt
```

```python
#loading profilers
stats1 = pstats.Stats('profile_stats1.prof')
stats2 = pstats.Stats('profile_stats2.prof')
stats3 = pstats.Stats('profile_stats3.prof')
```

```python
#obtain the metrics you want to compare, such as total time (tottime) or time per call (percall)
time_total = [stats1.total_tt, stats2.total_tt, stats3.total_tt]
time_per_call = [stats1.total_tt / stats1.total_calls,
                 stats2.total_tt / stats2.total_calls,
                 stats3.total_tt / stats3.total_calls]
```

```python
labels = ['Non-Parallel', 'Multiprocessing', 'MPI (mpi4py)']
```

```python
#Comparative chart
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(labels, time_total, label='Total Time', color='pink')


ax.set_xlabel('Mtodo de Clculo de Pi')
ax.set_ylabel('Tiempo (segundos)')
ax.set_title('Comparison of Execution Time by Pi Calculation Method with N=1000')
ax.legend()

plt.show()
```
